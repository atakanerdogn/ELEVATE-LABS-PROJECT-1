```python
# edtech in india

df_india = df[df['Country'] == 'India']
df_india
```

```python
df_india_edtech.describe()
```

```python
# indian edtech investment amount and valuation in usd

ieiav = df_india_edtech.groupby('Investment Amount (USD)')[['Valuation (USD)','Number of Investors']].sum().reset_index()
ieiav

ieiav.plot(kind='scatter', x='Investment Amount (USD)', y='Valuation (USD)')
plt.title('Investment Amount vs Valuation in indian ed-tech industry')
plt.xlabel('Investment Amount (USD)')
```

```python
 # indian edtech growth rate by age

 age = int(2025) - df_india_edtech['Year Founded']
 age

age_indian_edtech_growth = df_india_edtech.groupby(age)['Growth Rate (%)'].sum().reset_index()
age_indian_edtech_growth
```

```python
# avrage growth rate per industry per country

avg_growth_rate = df.groupby(['Country', 'Industry'])['Growth Rate (%)'].mean().reset_index()
avg_growth_rate

avg_growth_rate_germany = avg_growth_rate[avg_growth_rate['Country'] == 'Germany']
avg_growth_rate_germany
```

```python
#average gorwth rate for germany, different startups

avg_growth_rate_germany_ai = avg_growth_rate_germany[avg_growth_rate_germany['Industry'] == 'AI']
avg_growth_rate_germany_ai
```

```python

avg_growth_rate = df.groupby(['Country', 'Industry'])['Growth Rate (%)'].value_counts().reset_index()
avg_growth_rate

avg_growth_rate_germany = avg_growth_rate[avg_growth_rate['Country'] == 'Germany']
sorted_germany = avg_growth_rate_germany.sort_values(by='Growth Rate (%)', ascending=False).head()
sorted_germany.plot(kind='bar', x='Industry', y='Growth Rate (%)')
```

```python
# german top 5 ht with groth rate
germany = df[df['Country'] == 'Germany']
growth_rate_germany = germany.sort_values(by='Growth Rate (%)', ascending=False).value_counts().head()
growth_rate_germany
```

```python
# german healthtech top 5 by valuation valuecounts

healthtech_germany = germany[germany['Industry'] == 'HealthTech']
healthtech_germany.sort_values(by='Valuation (USD)', ascending=False).head()

sns.boxplot(x='Valuation (USD)', data=healthtech_germany)

```
