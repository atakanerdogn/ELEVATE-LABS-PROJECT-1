# AI Sector


```python
ger_ai = ger[ger['Industry'] == 'AI']
ger_ai.head()
```

```python
ger_ai.describe()
```

```python
ger_ai.info()
```

```python
ger_ai.count()
```

```python
# find top ai startups which get investment

ger_ai = ger_ai.sort_values(by='Investment Amount (USD)', ascending=False)
ger_ai.head()
```

```python
# top ai startups with growth rate

ger_ai = ger_ai.sort_values(by='Growth Rate (%)', ascending=False)
ger_ai.head()
```

```python
# top ai startups with number of investors

ger_ai = ger_ai.sort_values(by='Number of Investors', ascending=False)
ger_ai.head()
```

```python
# number of investors does not necessrily imply high investment
```

```python
# finding outliers

sns.boxplot(x=ger_ai['Growth Rate (%)'])
plt.show()
```

```python
# top ai startups by funding rounds

ger_ai = ger_ai.sort_values(by='Funding Rounds', ascending=False)
ger_ai.head(10)
```

```python
# startup_2027 which shows an interesting pattern

ger_ai = ger_ai.sort_values(by='Valuation (USD)', ascending=False)
ger_ai.head()
```

```python
# funding rounds and investment amounts

sns.barplot(x='Funding Rounds', y='Investment Amount (USD)', data=ger_ai, palette = 'crest')
plt.xticks(rotation=90)
plt.show()
```

```python
# funding rounds, investment amounts, valuation and investments

investment_general = ger.groupby(['Funding Rounds','Number of Investors', 'Valuation (USD)'])['Investment Amount (USD)'].sum()
investment_general
```

```python
ger_ai_num = ger_ai.select_dtypes(include=np.number)
ger_ai_num.corr().style.background_gradient(cmap='Blues')
```

```python
sns.heatmap(ger_ai_num.corr(),annot=True, cmap='Blues')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

x = ger_ai[['Valuation (USD)', 'Number of Investors', 'Year Founded', 'Funding Rounds', 'Growth Rate (%)']]
y = ger_ai['Investment Amount (USD)']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state = 42)

L = LinearRegression()
L.fit(x_train, y_train)

y_pred = L.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)

accuracy = L.score(x_test, y_test)
print("Accuracy", accuracy)
```

```python
sns.lineplot(x=y_test, y=y_pred, color='red')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
```
